```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import tweepy
import time
from datetime import datetime
# Twitter API Keys
from config import consumer_key, consumer_secret, access_token, access_token_secret

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()



# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())

```

```python
#Target Search 
target_terms = ("BBC", "CBS", "CNN",
                "FoxNews", "nytimes")
```

```python
sentiments_full = []

counter = 1
```

```python
for target in target_terms:

    compound_list = []
    positive_list = []
    negative_list = []
    neutral_list = []
    for x in range(5):

        public_tweets = api.user_timeline(target, page=x)
        for tweet in public_tweets:
        
            compound = analyzer.polarity_scores(tweet["text"])["compound"]
            pos = analyzer.polarity_scores(tweet["text"])["pos"]
            neu = analyzer.polarity_scores(tweet["text"])["neu"]
            neg = analyzer.polarity_scores(tweet["text"])["neg"]
            tweets_ago = counter
            
            compound_list.append(compound)
            positive_list.append(pos)
            negative_list.append(neg)
            neutral_list.append(neu)


            # Add sentiments for each tweet into an array
            sentiments_full.append({"User": target, 
                           "Source account": tweet["source"],
                           "Date": tweet["created_at"],        
                           "Text": tweet["text"],
                           "Compound": compound,
                           "Positive": pos,
                           "Negative": neu,
                           "Neutral": neg,
                           "Tweets Ago": counter})
            
            
            # Add to counter 
            counter = counter + 1
    
    

print(sentiments_full)
print("")
```

```python
#Convert sentiments to DataFrame
sentiments_data = pd.DataFrame.from_dict(sentiments_full)
sentiments_data
```

```python
sentiments_data = sentiments_data.set_index("User")
sentiments_data.head()
```

```python
sentiments_data.to_csv("news_sent_analysis_data.csv", encoding="utf-8", index=False)
```

```python
BBC = sentiments_data.loc["BBC", "Compound"]
CBS = sentiments_data.loc["CBS", "Compound"] 
CNN = sentiments_data.loc["CNN", "Compound"] 
Fox = sentiments_data.loc["FoxNews", "Compound"] 
nytimes = sentiments_data.loc["nytimes", "Compound"] 
BBC =BBC.reset_index()
CBS =CBS.reset_index()
CNN =CNN.reset_index()
Fox =Fox.reset_index()
nytimes =nytimes.reset_index()
```

```python
Compound_pd = pd.concat([BBC, CBS, CNN, Fox,nytimes],axis=1)
Compound_pd
Compound_pd.columns.values[1] = 'BBC Compound' 
Compound_pd.columns.values[3] = 'CBS Compound'
Compound_pd.columns.values[5] = 'CNN Compound'
Compound_pd.columns.values[7] = 'FoxNews Compound'
Compound_pd.columns.values[9] = 'NYtimes Compound'
Compound_pd.head()
```

```python
mpl_fig = plt.figure()
ax = mpl_fig.add_subplot(111)
ax.set_facecolor('pink')
BBC = ax.scatter(np.arange(len(Compound_pd["BBC Compound"])),
         Compound_pd["BBC Compound"], marker="o", color="blue", edgecolors="k",
         alpha=0.8, s=80 ) 
CBS = ax.scatter(np.arange(len(Compound_pd["BBC Compound"])),
         Compound_pd["CBS Compound"], marker="o", color="red", edgecolors="k",
         alpha=0.8,s=80)
CNN = ax.scatter(np.arange(len(Compound_pd["BBC Compound"])),
         Compound_pd["CNN Compound"], marker="o", color="orange", edgecolors="k",
         alpha=0.8 ,s=80)
Fox = ax.scatter(np.arange(len(Compound_pd["BBC Compound"])),
         Compound_pd["FoxNews Compound"], marker="o", color="green", edgecolors="k",
         alpha=0.8,s=80)
nytimes = ax.scatter(np.arange(len(Compound_pd["BBC Compound"])),
         Compound_pd["NYtimes Compound"], marker="o", color="black", edgecolors="k",
         alpha=0.8,s=80)
 
plt.title("Sentiment Analysis of News Media 10/15/18")
plt.ylabel("Tweet Polarity")
plt.xlabel("Tweets Ago")
plt.grid(True, linestyle='dashed')
plt.legend((BBC, CBS, CNN, Fox, nytimes), ("BBC Compound","CBS Compound","CNN Compound","FoxNews Compound","NYtimes Compound"),
           loc='lower left',
           bbox_to_anchor=(1, 0.5))
plt.savefig("Images/Sent_analysis_News_10_15")     
plt.show()
```

```python
avg_BBC = Compound_pd["BBC Compound"].mean()
avg_CBS = Compound_pd["CBS Compound"].mean()
avg_CNN = Compound_pd["CNN Compound"].mean()
avg_Fox = Compound_pd["FoxNews Compound"].mean()
avg_nytimes = Compound_pd["NYtimes Compound"].mean()

avg = {'Compound': pd.Series([avg_BBC, avg_CBS, avg_CNN, avg_Fox, avg_nytimes],index=['BBC','CBS','CNN','FoxNews','NYtimes'] )}
Compound_mean_df = pd.DataFrame(avg)
Compound_mean_df=Compound_mean_df.reset_index()
Compound_mean_df
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
x_axis = np.arange(0, len(Compound_mean_df["index"]))
tick_locations = []
for x in x_axis:
    tick_locations.append(x + 0.4)
ax.bar(x_axis, Compound_mean_df["Compound"], alpha=0.80, color="rbygk", align="edge")    

ax.set_ylabel('Tweet Polarity')
ax.set_title("Overrall News Sentiment based on Twitter (10-15-2018)")
ax.set_facecolor('lavender')
plt.xticks(tick_locations, Compound_mean_df["index"])
plt.axhline(0, color='k')
plt.savefig("Images/Overrall_News_Sentiment_Twitter_10_15")
plt.show()
```
